```python
import pandas as pd
import numpy as np
print("Pandas version:", pd.__version__)
```

```python
# Series (1D labeled array)
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
print("Series:")
print(s)
print("\nAccess by label:", s['c'])
print("Access by position:", s[0])
```

```python
# DataFrame (2D labeled structure)
data = {
    'name': ['Majed', 'Sara', 'Ahmed', 'Fatima'],
    'age': [18, 22, 25, 20],
    'city': ['Riyadh', 'Jeddah', 'Riyadh', 'Dammam']
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print("\nShape:", df.shape)
print("Columns:", df.columns.tolist())
print("Data types:\n", df.dtypes)
```

```python
# Inspect DataFrame
print("First 2 rows:")
print(df.head(2))
print("\nInfo:")
print(df.info())
print("\nSummary statistics:")
print(df.describe())
```

```python
# Select columns
print("Names column:")
print(df['name'])
print("\nAge and City:")
print(df[['age', 'city']])

# Select rows by position
print("\nFirst row:")
print(df.iloc[0])
print("\nRows 1-2:")
print(df.iloc[1:3])
```

```python
# Read CSV file (using the people.csv we created earlier)
df_csv = pd.read_csv('../people.csv')
print("DataFrame from CSV:")
print(df_csv)
print("\nData types:")
print(df_csv.dtypes)
```

```python
# Read CSV with custom options
df_csv2 = pd.read_csv('../people.csv', index_col=0)  # Use first column as index
print("With index_col:")
print(df_csv2)
```

```python
# Write DataFrame to CSV
df.to_csv('my_data.csv', index=False)  # index=False means don't save row numbers
print("✓ Saved DataFrame to my_data.csv")

# Read it back
df_read = pd.read_csv('my_data.csv')
print("\nRead back:")
print(df_read)
```
